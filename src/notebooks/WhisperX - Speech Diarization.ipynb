### WhisperX - Transcription and Diarization

```python
# Installing the WhisperX package from GitHub for transcription and diarization tasks.
!pip install git+https://github.com/m-bain/whisperx.git
```

```python
# Importing necessary modules. `whisperx` is for transcription and diarization, `gc` is for garbage collection.
%%time
import whisperx
import gc

# Setup: define the device for GPU acceleration, path to audio file, batch size, and computation precision.
device = "cuda"
audio_file = "/content/Audio_Data/Audio/CAR0001.mp3"
batch_size = 16
compute_type = "float16"

# Load and configure the WhisperX model for transcription with GPU and reduced precision for efficiency.
model = whisperx.load_model("large-v2", device, compute_type=compute_type)

# Load the audio file and transcribe it using the loaded model.
audio = whisperx.load_audio(audio_file)
result = model.transcribe(audio,batch_size=batch_size)
print(result["segments"])



# Load the English alignment model to enhance the accuracy of transcription alignment.
model_a, metadata = whisperx.load_align_model(language_code='en', device=device)
result = whisperx.align(result["segments"], model_a, metadata, audio, device, return_char_alignments=False)
print(result["segments"])


# Initialize the diarization pipeline with access token for API usage and configure for specific speaker detection.
# Note: The following code will require you to accept terms and conditions of the model in hugging face and put your own hugging face auth token in the pipeline.
diarize_model = whisperx.DiarizationPipeline(use_auth_token='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', device=device)
diarize_segments = diarize_model(audio)
diarize_model(audio, min_speakers=2, max_speakers=2)

# Combine diarization results with transcription to assign speaker labels to transcribed segments.
result = whisperx.assign_word_speakers(diarize_segments, result)
print(diarize_segments)
print(result["segments"])
```

```python
# Initialize a variable to track the first speaker detected in the audio segments.
first_speaker = None


# Loop through each segment of the transcribed and diarized result.
for segment in result["segments"]:
    # Retrieve the speaker identification and text from the current segment.
    speaker = segment["speaker"]
    text = segment["text"]

    # If the first speaker has not yet been identified, assign the current speaker as the first speaker.
    if first_speaker is None:
        first_speaker = speaker

    # If the current speaker is the same as the first speaker identified, label them as 'Doctor'.
    if speaker == first_speaker:
      speaker = 'Doctor'

    # Otherwise, label them as 'Patient'.
    else:
      speaker = 'Patient'

    # Print the start and end time of the segment along with the identified speaker label and the corresponding text.
    print(f"{segment['start']:06.1f}s {segment['end']:06.1f}s {speaker}: {text}")
```

```python
# Initialize an empty list to store formatted lines of dialogue, and a variable to track the first speaker.
lines = []
first_speaker = None

# Iterate over each segment from the transcription and diarization results.
for segment in result["segments"]:
    speaker = segment["speaker"]
    text = segment["text"]

    # Identify and assign the first speaker if not already done.
    if first_speaker is None:
        first_speaker = speaker

    # Assign the label 'Doctor' to the first speaker and 'Patient' to all other speakers.
    if speaker == first_speaker:
        speaker = 'Doctor'
    else:
        speaker = 'Patient'

    # Format the line with the speaker label and their dialogue, then add to the list.
    line = f"{speaker}: {text}"
    lines.append(line)

# Write the formatted dialogue lines to a text file.
with open("/content/output.txt", "w") as file:
    file.write("\n".join(lines))
```
